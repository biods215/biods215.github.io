One can use interactive visualization with Jupyter notebook.


```python
from sklearn import datasets

import plotly
import plotly.graph_objs as go
```

```python
plotly.offline.init_notebook_mode(connected=True)
```

## Sample dataset 

```python
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

## Scatter plot

```python
xaxis = 0
yaxis = 1
plotly.offline.iplot(
    go.Figure(
        data=[go.Scatter(
            x = iris.data[iris.target == target, xaxis],
            y = iris.data[iris.target == target, yaxis],
            name = iris.target_names[target],
            mode = 'markers',
            marker = dict(
                size = 10,
                line=dict(width=0.5),
                opacity=1,
            ),
        ) for target in set(iris.target)],               
        layout=go.Layout(
            hovermode = 'closest',
            margin=dict(
                l=0,
                r=0,
                b=0,
                t=-10,
            ),
            xaxis= dict(
                title= iris.feature_names[xaxis],
            ),
            yaxis=dict(
                title= iris.feature_names[yaxis],
            ),
            title=iris.DESCR.splitlines()[0],
        )
    ), 
    filename='iris-scatter'
)
```

## Scatter plot (3d)

```python
xaxis = 0
yaxis = 1
zaxis = 2
plotly.offline.iplot(
    go.Figure(
        data=[go.Scatter3d(
            x = iris.data[iris.target == target, xaxis],
            y = iris.data[iris.target == target, yaxis],
            z = iris.data[iris.target == target, zaxis],            
            name = iris.target_names[target],
            mode = 'markers',
            marker = dict(
                size = 10,
                line=dict(width=0.5),
                opacity=1,
            ),
        ) for target in set(iris.target)],               
        layout=go.Layout(
            hovermode = 'closest',
            margin=dict(
                l=0,
                r=0,
                b=0,
                t=-10,
            ),
            title=iris.DESCR.splitlines()[0],
        )
    ), 
    filename='iris-scatter'
)
```
